## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data["Age Group"] = pd.cut(x=purchase_data["Age"], bins=[0,9, 14, 19, 24,29,34,39,111])

ages = purchase_data.groupby("Age Group")["SN"].nunique()

ages_percent = round(ages/ages.sum() *100, 2)

ages_df = pd.DataFrame({"Age Group Count": ages,
                        "Age by Percent": ages_percent})
ages_df

#0,9, 14, 19, 24,29,34,39,111
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(x=purchase_data["Age"], bins=[0,9, 14, 19, 24,29,34,39,111])

purchase_count_age = purchase_data.groupby("Age Group")["Purchase ID"].count()


#average purchase price by age
average_price_age = round(purchase_data.groupby("Age Group")["Price"].mean(), 2)
average_price_age

#total purchases by age
total_purchase_age = purchase_data.groupby("Age Group")["Price"].sum()
total_purchase_age

#create df 
gender_df = pd.DataFrame({"Purchase Count by Age Group": purchase_count_age,
                          "Avg Price by Age Group": average_price_age, 
                          "Total Purchases by Age Group": total_purchase_age})
gender_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#total purchase value + sort by SN 
whale_count = purchase_data.groupby("SN")["Price"].sum()
whale_count_max = (whale_count).sort_values(ascending=False)
whale_count_max.head(5)

#purchase count
whale_purchase_count = purchase_data.groupby("SN")["Purchase ID"].count()
whale_purchase_count_max = (whale_purchase_count).sort_values(ascending=False)
whale_purchase_count_max.head(5)

#average purchase price    total price spent per person as average 
whale_average_price = round(purchase_data.groupby("SN")["Price"].mean(), 2)
whale_average_price

#descending by count

whale_df = pd.DataFrame({"Whale Total Spent": whale_count,
                          "Whale Purchase Count": whale_purchase_count,
                          "Whale Average Purchase Price": whale_average_price})

whale_df_1 = whale_df.sort_values(by=["Whale Total Spent"], ascending=False)
whale_df_1.head(5)
```

## Most Popular Items

* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

popular_build = pd.DataFrame(purchase_data, columns=["Item ID","Item Name","Price"])
popular_build.head(5)

purchase_item_build = round(purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"], 2)


#works
purchase_value_build = popular_build.groupby(["Item ID","Item Name"])["Price"].sum()

#works
purchase_count_build = popular_build.groupby(["Item ID"])["Item Name"].value_counts()


popular_df = pd.DataFrame({"Purchase Count": purchase_count_build,
                             "Average Price": purchase_item_build,
                             "Total Purchase Value": purchase_value_build})
popular_df.head(5)



popular_df_1 = popular_df.sort_values(by=["Purchase Count"], ascending=False)
popular_df_1.head(5)

```
